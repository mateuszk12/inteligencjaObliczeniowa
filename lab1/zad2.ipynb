```python
import pandas as pa
import matplotlib.pyplot as plt
miasta = pa.read_csv("miasta.csv")
print(miasta.values)
```

```python
miasta.loc[len(miasta.index)] = [2010,460,555,405]
```

```python
plt.ylabel("Liczba ludnosci w tys.")
plt.xlabel("Lata")
plt.title("Liczba ludnosci Gdanska")
plt.plot(miasta["Rok"],miasta["Gdansk"],color="red",marker="o")
plt.show()
```

```python
plt.ylabel("Liczba ludnosci w tys.")
plt.xlabel("Lata")
plt.title("Liczba ludnosci Gdanska")
plt.plot(miasta["Rok"],miasta["Gdansk"],color="blue",marker="o",label="Gdansk")
plt.plot(miasta["Rok"],miasta["Poznan"],color="red",marker="o",label="Poznan")
plt.plot(miasta["Rok"],miasta["Szczecin"],color="green",marker="o",label="Szczecin")
plt.legend(loc="upper left")
plt.show()
```

```python
miasta["Gdansk"]=miasta["Gdansk"]-miasta["Gdansk"].mean()/miasta["Gdansk"].std()
miasta["Poznan"]=miasta["Poznan"]-miasta["Poznan"].mean()/miasta["Poznan"].std()
miasta["Szczecin"]=miasta["Szczecin"]-miasta["Szczecin"].mean()/miasta["Szczecin"].std()
print(miasta)
```

```python
miasta["Gdansk"]=(miasta["Gdansk"]-miasta["Gdansk"].min())/(miasta["Gdansk"].max()-miasta["Gdansk"].min())
miasta["Szczecin"]=(miasta["Szczecin"]-miasta["Szczecin"].min())/(miasta["Szczecin"].max()-miasta["Szczecin"].min())
miasta["Poznan"]=(miasta["Poznan"]-miasta["Poznan"].min())/(miasta["Poznan"].max()-miasta["Poznan"].min())
print(miasta)
```
